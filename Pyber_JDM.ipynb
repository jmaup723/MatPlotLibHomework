```python
%matplotlib inline
# # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# # # File to Load (Remember to change these)
city_data_to_load = "Pyber/data/city_data.csv"
ride_data_to_load = "Pyber/data/ride_data.csv"

# # # # Read the City and Ride Data
city_file = pd.read_csv(city_data_to_load)
ride_file = pd.read_csv(ride_data_to_load)


# # # Combine the data into a single dataset
city_ride = pd.merge(city_file, ride_file, how="left", on=["city"])
# # # Display the data table for preview
city_ride
# # # ride_file.head()
del city_ride['date']
city_ride
```

```python
bycity = city_ride.groupby('city')
avefare = bycity.mean()
totaldriver = bycity.count()
# # totaldriver = bycity['driver_count'].nunique()
# totalride = bycity['ride_id'].count()
avefare
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# x_limit = 40

city_ride_rural = city_ride.loc[city_ride["type"] == "Rural"]
rural_bycity = city_ride_rural.groupby('city')
rural_ave = rural_bycity.mean()
rural_driver = rural_bycity.count()

city_ride_urban = city_ride.loc[city_ride["type"] == "Urban"]
urban_bycity = city_ride_urban.groupby('city')
urban_ave = urban_bycity.mean()
urban_driver = urban_bycity.count()

city_ride_suburban = city_ride.loc[city_ride["type"] == "Suburban"]
suburban_bycity = city_ride_suburban.groupby('city')
suburban_ave = suburban_bycity.mean()
suburban_driver = suburban_bycity.count()


rural = plt.scatter(rural_driver['driver_count'], rural_ave['fare'], s=rural_ave['driver_count'], alpha=1.0, marker="o", facecolors="gold", edgecolors="black")
urban = plt.scatter(urban_driver['driver_count'], urban_ave['fare'] , s=urban_ave['driver_count'], alpha=1.0,marker="o", facecolors="coral", edgecolors="black")
suburban = plt.scatter(suburban_driver['driver_count'], suburban_ave['fare'], s=suburban_ave['driver_count'], alpha=1.0, marker="o", facecolors="lightskyblue", edgecolors="black")

# Build the scatter plots for each city types
# plt.scatter(avefare['fare'], totaldriver['driver_count'], marker="o", facecolors="red", edgecolors="black")
plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Ride Per City")
plt.ylabel("Average Fare ($)")
plt.legend(handles=[rural, urban, suburban], loc="upper right")
plt.grid()
# Incorporate the other graph properties
plt.text(68, 40, 'Circle Size represents driver count per city', horizontalalignment='right', verticalalignment='center')
# Create a legend

# Incorporate a text label regarding circle size

```
